```python
import time
import sys

# ===== CORES ANSI =====
RESET = "\033[0m"
VERDE = "\033[1;32m"
VERMELHO = "\033[1;31m"
AMARELO = "\033[1;33m"
AZUL = "\033[1;34m"
CINZA = "\033[1;30m"

temperatura_interna = 22.3
temperatura_externa = 50.6
integridade_estrutural = 50
nivel_energia = [99, 97, 94, 95]
pressao_tanques = 3973
modulos_criticos = {
    "IMU": "Online",
    "GNC": "Online",
    "RCS": "Online",
    "FTS": "Online",
    "COMMS": "Online"
}
status_lancamento = True

def escrever_lento(mensagem, velocidade, end='\n'):
    mensagem += end
    for i in mensagem:
        delay = velocidade / 1000
        sys.stdout.write(i)
        sys.stdout.flush()
        time.sleep(delay)


def separador(titulo=""):
    linha = "=" * 50
    if titulo:
        print(f"\n{AZUL}{linha}")
        print(f"{titulo.center(50)}")
        print(f"{linha}{RESET}")
    else:
        print(f"{AZUL}{linha}{RESET}")


class Verificacoes:

    @staticmethod
    def temperatura(interna, externa):
        erros_temperatura = []
        if interna > 30 or interna < 15:
            erros_temperatura.append(f'{VERMELHO}Temperatura interna fora da margem!\nTEMP. INTERNA: {interna}°C | ESPERADO: 15-30°C{RESET}')
        if externa > 50 or externa < -50:
            erros_temperatura.append(f'{VERMELHO}Temperatura externa fora da margem!\nTEMP. EXTERNA: {externa}°C | ESPERADO: (-50)-50°C{RESET}')
        return erros_temperatura
    
    @staticmethod
    def integridade(integridade):
        erros_integridade = []
        if integridade < 100:
            erros_integridade.append(f'{VERMELHO}Integridade estrutural comprometida! Microfissuras detectadas.{RESET}')
        return erros_integridade
    
    @staticmethod
    def energia(status_baterias):
        erros_energia = []
        baterias = {
            "Bateria 1": status_baterias[0],
            "Bateria 2": status_baterias[1],
            "Bateria 3": status_baterias[2],
            "Bateria 4": status_baterias[3],
        }
        nivel_total = sum(status_baterias) / 4
        if nivel_total < 95:
            erros_energia.append(f'{AMARELO}Energia total: {nivel_total}%{RESET}')
        for bateria in baterias:
            if baterias[bateria] < 90:
                erros_energia.append(f"{VERMELHO}{bateria} abaixo da margem segura! ({baterias[bateria]}%){RESET}")
        return erros_energia
    
    @staticmethod
    def pressao(pressao):
        erros_pressao = []
        if pressao > 4500 or pressao < 3000:
            erros_pressao.append(f'{VERMELHO}Pressão fora do permitido!\nAtual: {pressao} PSI | Esperado: 3000-4500 PSI{RESET}')
        return erros_pressao
    
    @staticmethod
    def modulos(modulos):
        erros_modulo = []
        for modulo in modulos:
            if modulos[modulo] != "Online":
                erros_modulo.append(f"{VERMELHO}{modulo}: OFFLINE{RESET}")
        return erros_modulo


def main():
    cont_erros = 0
    erros = []

    separador("SISTEMA DE PRÉ-LANÇAMENTO")
    escrever_lento(f"\n{CINZA}Inicializando sistemas...{RESET}", 50)
    time.sleep(1)

    # TEMPERATURA
    separador("CHECAGEM DE TEMPERATURA")
    escrever_lento("\nVerificando sensores térmicos...\n", 50)
    time.sleep(1)
    erros.append(Verificacoes.temperatura(temperatura_interna, temperatura_externa))
    if erros[cont_erros]:
        print(f"{VERMELHO}[FALHA]{RESET}")
        for erro in erros[cont_erros]:
            print(erro)
        status_lancamento = False
    else:
        print(f"{VERDE}[OK]{RESET}")
    cont_erros += 1

    # INTEGRIDADE
    separador("INTEGRIDADE ESTRUTURAL")
    escrever_lento("\nAnalisando casco e estrutura...\n", 50)
    time.sleep(1)
    erros.append(Verificacoes.integridade(integridade_estrutural))
    if erros[cont_erros]:
        print(f"{VERMELHO}[FALHA]{RESET}")
        for erro in erros[cont_erros]:
            print(erro)
        status_lancamento = False
    else:
        print(f"{VERDE}[OK]{RESET}")
    cont_erros += 1

    # ENERGIA
    separador("SISTEMA DE ENERGIA")
    escrever_lento("\nVerificando baterias...\n", 50)
    time.sleep(1)
    erros.append(Verificacoes.energia(nivel_energia))
    if erros[cont_erros]:
        print(f"{AMARELO}[ALERTA]{RESET}")
        for erro in erros[cont_erros]:
            print(erro)
        status_lancamento = False
    else:
        print(f"{VERDE}[OK]{RESET}")
    cont_erros += 1

    # PRESSÃO
    separador("PRESSÃO DOS TANQUES")
    escrever_lento("\nMonitorando pressão...\n", 50)
    time.sleep(1)
    erros.append(Verificacoes.pressao(pressao_tanques))
    if erros[cont_erros]:
        print(f"{VERMELHO}[FALHA]{RESET}")
        for erro in erros[cont_erros]:
            print(erro)
        status_lancamento = False
    else:
        print(f"{VERDE}[OK]{RESET}")
    cont_erros += 1

    # MÓDULOS
    separador("MÓDULOS CRÍTICOS")
    escrever_lento("\nChecando status dos módulos...\n", 50)
    time.sleep(1)
    erros.append(Verificacoes.modulos(modulos_criticos))
    if erros[cont_erros]:
        print(f"{VERMELHO}[FALHA]{RESET}")
        for erro in erros[cont_erros]:
            print(erro)
        status_lancamento = False
    else:
        print(f"{VERDE}[OK]{RESET}")
    cont_erros += 1

    # RESULTADO FINAL
    separador("STATUS FINAL")
    if status_lancamento:
        print(f"{VERDE}\n✔ TODOS OS SISTEMAS OPERACIONAIS{RESET}")
        print(f"{VERDE}🚀 AUTORIZADO PARA LANÇAMENTO\n{RESET}")
    else:
        print(f"{VERMELHO}\n✖ FALHAS DETECTADAS{RESET}")
        print(f"{VERMELHO}🚫 LANÇAMENTO ABORTADO\n{RESET}")


if __name__ == "__main__":
    main()
```
